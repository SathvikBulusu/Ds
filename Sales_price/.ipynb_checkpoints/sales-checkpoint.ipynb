### Given below is the Encoding for categorical varaibles ( 43 )  for building a predictive model 

### Train Dataset Encoding


```python
cat_col = x.select_dtypes(include=['object']).columns

enco = OneHotEncoder(drop='first',sparse_output=False ,handle_unknown ='ignore')

encoded_data = enco.fit_transform(x[cat_col])

encoded_columns = enco.get_feature_names_out(cat_col)

cat_encoded_data = pd.DataFrame(encoded_data,columns = encoded_columns,index = x.index)

x_encoded = x.drop(columns=cat_col)

x_encoded = pd.concat([x_encoded,cat_encoded_data], axis=1)

print(x_encoded.info())

print(x_encoded.head())
```

```python
x_encoded # this is the encoded full training dataset 
```

```python
x_encoded.drop_duplicates() ## remember to drop the duplicates next time 
```

### Test Dataset Encoding 

```python
categorical_encoding = y.select_dtypes(include=['object']).columns

test_encoding = enco.fit_transform(y[categorical_encoding])

test_encoded_columns = enco.get_feature_names_out(categorical_encoding)

test_encoded_data = pd.DataFrame(test_encoding,columns=test_encoded_columns, index=y.index)

y_encoded = y.drop(columns=categorical_encoding)

y_encoded = pd.concat([y_encoded,test_encoded_data],axis =1)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
model = LinearRegression()

xx = x_encoded.drop(columns=['SalePrice'])
yy = x_encoded['SalePrice']



x_train , x_test , y_train , y_test = train_test_split(xx,yy,test_size=0.2,random_state = 42)

y_train = np.log1p(y_train)


model.fit(x_train,y_train)
```

```python
# Find missing columns in test dataset
missing_cols = set(x_train.columns) - set(y_encoded.columns)
extra_cols = set(y_encoded.columns) - set(x_train.columns)

print(f"Missing in test: {missing_cols}")
print(f"Extra in test: {extra_cols}")
```

```python
# Ensure all missing columns are added to test data
for col in missing_cols:
    y_encoded[col] = 0  # Add missing columns with 0 values

# Ensure the column order is the same
y_encoded = y_encoded[x_train.columns]
```

```python
y_pred = model.predict(x_test)

y_pred_log = np.expm1(y_pred)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test,y_pred_log)
mse = mean_squared_error(y_test,y_pred_log)
r = r2_score(y_test,y_pred_log)

print(f"Mean absolute error : {mae:.2f}")
print(f"Mean Squared Error : {mse:.2f}")
print(f"r2 : {r:.2f}")
```

```python
test_predictions = model.predict(y_encoded)

test_predictions = np.where(test_predictions < 0, 0, test_predictions)

# Convert to integer prices
test_predictions = np.round(test_predictions, 0).astype(int)  

submission = pd.DataFrame({'Id':y_encoded.index , 'SalePrice':test_predictions})

submission.to_csv("submission.csv",index =False)
```
