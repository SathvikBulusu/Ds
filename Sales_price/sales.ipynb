### Problem Statement :- Predict the Sales prices 

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
import matplotlib.pyplot 
import seaborn as sns 
```

--------------------------------------------------------------------

### Data science Stage 2 :- Data Collection ( stage 1 is understanding the problem / need / requirement ) 

```python
train_data = pd.read_csv("train.csv")

train_data
```

------------------------------------------------------------------------------------------


### Data Science Stage 3 :- Data Preprocessing ( i am taking the liberty of adding data cleaning in this stage ) 

```python
for col in train_data.columns: # imputation method for train dataset
    if train_data[col].dtype in ['int64','float64']:
        train_data[col] = train_data[col].fillna(train_data[col].median())
    else:
        train_data[col] = train_data[col].fillna(train_data[col].mode()[0])
```

```python
test_data = pd.read_csv("test.csv")

test_data
```

```python
for col in test_data.columns:# imputation method for test dataset 
    if test_data[col].dtype in ['int64','float64']:
        test_data[col] = test_data[col].fillna(test_data[col].median())
    else:
        test_data[col] = test_data[col].fillna(test_data[col].mode()[0])
        
```

```python
x = train_data.drop(columns='SalePrice')
y = train_data['SalePrice']
```

### Given below is the Encoding for categorical varaibles ( 43 )  for building a predictive model 

### Train Dataset Encoding


```python
cat_col = x.select_dtypes(include=['object']).columns
num_col = x.select_dtypes(include=['int64','float64']).columns

enco = OneHotEncoder(drop='first',sparse_output=False ,handle_unknown ='ignore')

encoded_data = enco.fit_transform(x[cat_col])

encoded_columns = enco.get_feature_names_out(cat_col)

cat_encoded_data = pd.DataFrame(encoded_data,columns = encoded_columns,index = x.index)

x_encoded = x.drop(columns=cat_col)

x_encoded = pd.concat([x[num_col],cat_encoded_data], axis=1)

print(x_encoded.info())

print(x_encoded.head())
```

```python
x_encoded # this is the encoded full training dataset 
```

### Test Dataset Encoding 

```python
test_encoded_data = enco.transform(test_data[cat_col])

test_encoded_columns = enco.get_feature_names_out(cat_col)

y_encoded = pd.DataFrame(test_encoded_data,columns=test_encoded_columns, index=test_data.index)

y_encoded = pd.concat([test_data[num_col],y_encoded],axis=1)
```

```python
y_encoded.info()
y_encoded.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
scaler = StandardScaler()

x_scaled = pd.DataFrame(scaler.fit_transform(x_encoded),columns = x_encoded.columns, index = x_encoded.index)
y_scaled = pd.DataFrame(scaler.fit_transform(y_encoded),columns = y_encoded.columns, index = y_encoded.index)

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state =42)

y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

model = LinearRegression()

model.fit(x_train,y_train_log)
```

```python
y_pred_log= model.predict(x_test)
max_log_val = np.log(1e6)
min_log_val = np.log(1e4)
y_pred_log= np.clip(y_pred_log,min_log_val,max_log_val)
y_pred = np.expm1(y_pred_log)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r = r2_score(y_test,y_pred)
rmsle = np.sqrt(mean_squared_error(y_test_log,y_pred_log))

print(f"Mean absolute error : {mae:.2f}")
print(f"Mean Squared Error : {mse:.2f}")
print(f"r2 : {r:.2f}")
print(rmsle)
```

```python
test_predictions = model.predict(y_scaled)

test_predictions = np.clip(test_predictions,min_log_val,max_log_val)

test_pred = np.expm1(test_predictions)

submission = pd.DataFrame({'Id':test_data['Id'] , 'SalePrice':test_pred})

submission.to_csv("submission.csv",index =False)
```
