### Outlier Detection

```python
def detect_outliers(train_dataset,column):
    q1=train_dataset[column].quantile(0.25)
    q3=train_dataset[column].quantile(0.75)
    IQR = q3-q1
    lower_bound = q1-1.5*IQR
    upper_bound = q3+1.5*IQR
    outliers = train_dataset[(train_dataset[column] < lower_bound)| (train_dataset[column] >upper_bound)][column]
    return len(outliers),lower_bound,upper_bound

outlier_sum={}
for col in train_dataset:
    count,lower,upper = detect_outliers(train_dataset,col)
    outlier_sum[col] = {'outlier_count' :count , 'Lower_Bound' : lower, "Upper_Bound" : upper}
    
outlier_df = pd.DataFrame(outlier_sum).T
outlier_df = outlier_df.sort_vakyes("outlier_count",ascending = False)
print(outlier_df.head(10))
print("Columns with outliers : ",len(outlier_df[outlier_df['outlier_count']>0]))
print("maximum outliers in a column : ",outkuer_df['outlier_count'].max())
print(" Percentage of outliers in top column :",(outlier_df['outlier_count'].max()/ train_dataset.shape[0])*100, '%')
```
